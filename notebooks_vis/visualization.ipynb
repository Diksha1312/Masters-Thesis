```python
DIR = "artifacts"
MODEL_DIR = os.path.join("models", "LSTMModel", "nsp1", "lightning_logs", "version_0")
file_paths = [
    os.path.join(DIR, "nsp1_mutation_rate.csv"),
    os.path.join(DIR, "nsp3_mutation_rate.csv"),
    os.path.join(DIR, "nsp13_mutation_rate.csv"),
    os.path.join(DIR, "nsp15_mutation_rate.csv")
]
```

```python
MODEL_DIR
```

```python


# Calculate column sums for each file
column_sums_list = [pd.read_csv(file_path).sum() for file_path in file_paths]

# Create a 2x2 grid of pie charts
fig, axes = plt.subplots(2, 2, figsize=(12, 12))

for i, (file_path, column_sums) in enumerate(zip(file_paths, column_sums_list)):
    # Calculate the subplot position
    row = i // 2
    col = i % 2

    # Plot the pie chart in the specified subplot
    axes[row, col].pie(column_sums, labels=column_sums.index, autopct='%1.1f%%', startangle=90)
    axes[row, col].set_title(f'Pie Chart for Column Sums - {os.path.basename(file_path)}')

# Adjust layout to prevent overlap
plt.tight_layout()
# Save the figure
# os.makedirs("Figures", exist_ok=True)
# plt.savefig('Figures/pie_charts_proteins.png')

# Show the plot
plt.show()
```

```python
nucleotide_colors = {'AC': 'skyblue', 'AT': 'salmon', 'AG': 'lightgreen', 'CA': 'lightcoral',
                     'CT': 'gold', 'CG': 'mediumorchid', 'TA': 'cornflowerblue', 'TC': 'tomato',
                     'TG': 'palegreen', 'GA': 'rosybrown', 'GC': 'deepskyblue', 'GT': 'thistle'}

# Calculate column sums for each file
column_sums_list = [pd.read_csv(file_path).sum() for file_path in file_paths]

# Create a 2x2 grid of bar plots
fig, axes = plt.subplots(2, 2, figsize=(12, 12))

for i, (file_path, column_sums) in enumerate(zip(file_paths, column_sums_list)):
    # Calculate the subplot position
    row = i // 2
    col = i % 2

    # Plot the bar plot in the specified subplot with consistent colors
    sns.barplot(x=column_sums.index, y=column_sums.values, ax=axes[row, col],
                palette=[nucleotide_colors[n] for n in column_sums.index])
    
    axes[row, col].set_title(f'Bar Plot for Column Sums - {os.path.basename(file_path)}')
    axes[row, col].set_xlabel('Nucleotides')
    axes[row, col].set_ylabel('Sum of Mutation Rates')

# Adjust layout to prevent overlap
plt.tight_layout()

os.makedirs("Figures", exist_ok=True)
# Save the figure
plt.savefig('Figures/bar_plots_proteins.png')

# Show the plot
plt.show()
```

```python

def plot_metrics(ax, model_name, nsp_name, version_name):
    MODEL_DIR = os.path.join("models", model_name, nsp_name, "lightning_logs", version_name)
    file_path = os.path.join(MODEL_DIR, "metrics.csv")
    train_val_loss_df = pd.read_csv(file_path)
    del train_val_loss_df['step']
    del train_val_loss_df['test_loss']
    train_val_loss_df.set_index("epoch", inplace=True)
    train_val_loss_df.dropna(axis=1, how="all", inplace=True)
    train_val_loss_df.dropna(axis=0, how="all", inplace=True)
    sns.lineplot(data=train_val_loss_df, ax=ax)

# Create a figure with 1 row and 2 columns
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Plot for GRUModel - nsp1
plot_metrics(axes[0], "LSTMModel", "nsp15", "version_0")
axes[0].set_title('LSTMModel - nsp15')

# Plot for LSTMModel - nsp15
plot_metrics(axes[1], "LSTMModel", "nsp15", "version_3")
axes[1].set_title('LSTMModel_IncreasedHiddenSize_MoreFCs - nsp15')

# Adjust layout for better spacing
plt.tight_layout()

os.makedirs("figures", exist_ok=True)
# Save the figure
plt.savefig("figures/default_lstm__MoreHiddenSize_moreFCs_50-30-20split__nsp15.png")
# Show the plots
plt.show()
```
